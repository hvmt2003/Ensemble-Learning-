```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv('/content/placement.csv')
```

```python
dataset.head()
```

```python
x=dataset.iloc[:,0:-1]
y=dataset.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

Selected models are as follows:
1.Linear Regression
2.Decision Tree Regression
3.Support Vector Machine Classifiers

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_train,y_train),lr.score(x_test,y_test)
```

```python
dtr=DecisionTreeRegressor()
dtr.fit(x_train,y_train)
```

```python
dtr.score(x_train,y_train),dtr.score(x_test,y_test)
```

```python
svr=SVR()
svr.fit(x_train,y_train)
```

```python
svr.score(x_train,y_train),svr.score(x_test,y_test)
```

Applying Voting Regressor

```python
from sklearn.ensemble import VotingRegressor
```

```python
vr=VotingRegressor([('lr',lr),('dtr',dtr),('svr',svr)])
vr.fit(x_train,y_train)
```

```python
vr.score(x_train,y_train),vr.score(x_test,y_test)
```

```python
df={'lr':lr.predict(x_test),'dt':dtr.predict(x_test),'svr':svr.predict(x_test),'vr':vr.predict(x_test)}
```
