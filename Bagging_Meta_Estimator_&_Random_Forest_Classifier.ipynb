```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
```

```python
x,y=make_moons(n_samples=1000,noise=0.15)
```

```python
df={'x1':x[:,0],'x2':x[:,1],'y':y}
dataset=pd.DataFrame(df)
```

```python
sns.scatterplot(x='x1',y='x2',hue='y',data=dataset)
plt.show()
```

```python
x_a=dataset.iloc[:,:-1]
y_a=dataset.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_a,y_a,test_size=0.20,random_state=42)
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
sv=SVC()
sv.fit(x_train,y_train)
```

```python
sv.score(x_train,y_train),sv.score(x_test,y_test)
```

```python
bg=BaggingClassifier(estimator=SVC(),n_estimators=100)
bg.fit(x_train,y_train)
```

```python
bg.score(x_train,y_train),bg.score(x_test,y_test)
```

```python
rf=RandomForestClassifier(n_estimators=100)
rf.fit(x_train,y_train)
```

```python
rf.score(x_train,y_train),rf.score(x_test,y_test)
```
